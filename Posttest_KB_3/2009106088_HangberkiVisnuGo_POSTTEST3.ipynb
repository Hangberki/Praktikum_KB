```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("music_genre.csv")

df.head(10)
```

# GRAFIK MENUNJUKAN POPULARITAS GENRE MUSIK

```python
from numpy import size


music_genre= df.dropna(subset=['music_genre']).sort_values(by=['music_genre'])['music_genre'].unique()

popular = df.groupby(['music_genre'])['popularity'].sum()

plt.figure(figsize = (30,15))
plt.plot(music_genre, popular, linestyle='--', marker='o', color='red')
plt.grid()
plt.title('Tingkat popularitas genre musik')
plt.xlabel('genre', fontsize=12)
plt.ylabel('popularitas', fontsize=12)
plt.fill_between(music_genre, popular, color='lightgreen')

plt.show()
```

# GRAFIK MENUNJUKKAN TIPE YANG DI MAINKAN OLEH MUSISI 

```python
tipe = df.dropna(subset=['mode']).sort_values(by=['mode'])['mode'].unique()

total = df.groupby(['mode'])['artist_name'].count()

patches, texts, pcts = plt.pie(total, labels=tipe, autopct='%.1f%%', wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})


plt.title('Persentase tipe yang digunakan seorang artist')
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```

# GRAFIK MENUNJUKAN PERBANDINGAN MODE SESUAI GENRE

```python
sns.set_style("ticks")
plt.figure(figsize = (25,9))
sns.countplot(data = df, x = "music_genre", hue = "mode")
plt.title('Perbandingan mode pada stiap genre musik')
```
