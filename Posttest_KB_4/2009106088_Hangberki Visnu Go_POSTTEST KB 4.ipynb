# Import Library dan Read Data
Import library.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("music_genre.csv")
df.head(5)
```

# Data Splitting
Melakukan subset target dan feature

```python
X = df.iloc[:,:-1] # Target
y = df.iloc[:,-1] # Feature
```

Buat pembagian train_test_split dengan perbandinga 70:30. Parameter test_size yang mengatur berapa besar pembagian diberi nilai 0.3 karena 0.3 sama dengan 30%

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Menampilkan dimensi hasil splitting

```python
print("Dimensi data\t:", df.shape)
print()
print("Dimensi X_train\t:", X_train.shape)
print("Dimensi X_test\t:", X_test.shape)
print("Dimensi y_train\t:", y_train.shape)
print("Dimensi y_test\t:", y_test.shape)
```

Normalisasi Data: Min-Max Scaler
Import library untuk MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

Membuat kelas MinMaxScaler ke variabel min_max_scaler.

```python
min_max_scaler = MinMaxScaler()
```

Menduplikat DataFrame df ke to_be_normalized, kemudian menampilkan kolom-kolomnya.

```python
to_be_normalized = df.copy()
to_be_normalized.columns
```

Melakukan proses normalisasi data pada kolom "loudness", "duration_ms", "energy"

```python
x_scaler = min_max_scaler.fit_transform(to_be_normalized[["loudness", "duration_ms", "energy"]])
```

Mengonversi hasil normalisasi menjadi DataFrame

```python
df_normalized = pd.DataFrame(x_scaler)
df_normalized
```

Menampilkan nilai min dan max dari DataFrame hasil normalisasi untuk membuktikan bahwa normalisasi dieksekusi dengan benar. Hasil normalisasi memiliki rentang [0, 1].

```python
print("Nilai min:")
print(df_normalized.min())
print()
print("Nilai max:")
print(df_normalized.max())
```

 # Data Cleaning: null

```python
from sklearn.impute import SimpleImputer
```

```python
imputer_mean = SimpleImputer(strategy="mean")
imputer_most_frequent = SimpleImputer(strategy="most_frequent")
```

```python
df.isna().sum()
```

```python
df["instance_id"] = imputer_mean.fit_transform(df[["instance_id"]])
df["track_name"] = imputer_most_frequent.fit_transform(df[["track_name"]])
df["popularity"] = imputer_most_frequent.fit_transform(df[["popularity"]])
df["acousticness"] = imputer_most_frequent.fit_transform(df[["acousticness"]])
df["danceability"] = imputer_most_frequent.fit_transform(df[["danceability"]])
df["duration_ms"] = imputer_mean.fit_transform(df[["duration_ms"]])
df["energy"] = imputer_most_frequent.fit_transform(df[["energy"]])
df["instrumentalness"] = imputer_most_frequent.fit_transform(df[["instrumentalness"]])
df["key"] = imputer_most_frequent.fit_transform(df[["key"]])
df["liveness"] = imputer_most_frequent.fit_transform(df[["liveness"]])
df["loudness"] = imputer_most_frequent.fit_transform(df[["loudness"]])
df["speechiness"] = imputer_most_frequent.fit_transform(df[["speechiness"]])
df["tempo"] = imputer_most_frequent.fit_transform(df[["tempo"]])
df["obtained_date"] = imputer_most_frequent.fit_transform(df[["obtained_date"]])
```

```python
df.isna().sum()
```
